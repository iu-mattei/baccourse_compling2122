```python
import re #импортируем библиотеку для использования регулярных выражений
def prep(text):
#можно сохранить разбивку на предложения  
    text = re.sub('[^.!? \w]', ' ', text) #превратим все нечислобуквенные символы кроме точки, восклицательного и вопросительного знаков, пробела в пробелы
    text = re.sub('\s+', ' ', text) #удалим повторяющиеся пробелы  
    sentences = re.split(r'[\.?!][ ]?', text) #при делении предложений "доверяем" не удаленным раньше знакам препинания, предполагаем, что может не оказаться нужных пробелов

#очистка до текстовых символов (нормализация регистра до нижнего, удаление оставшихся знаков препинания и чисел)  
    text = text.lower()
    text = re.sub('[^а-я ]', '', text)
    cleanedText = re.sub('\s+', ' ', text)
    
#разделим строку на слова по пробелу
    tokens = re.split(r'[ ]', cleanedText) #список с токенами
    
    #длина текста (количество слов)
    length = len(tokens)

#удалим стоп-слова и посчитаем их долю от начальной длины текста    
    stopwords = ['в', 'к', 'до', 'за', 'при', 'не'] #определяем стоп-слова
    textlist = cleanedText.split()
    nlst = []
    a = 0
    for x in textlist: 
        if x in stopwords:
            continue
        nlst.append(x) #добавляем все слова, кроме шумовых, в новый список
        a += 1
    ststr = ' '.join(nlst)
    stopout = re.split(r'[ ]', ststr)

    sw = round(len(stopwords)/length, 2) #долю стоп-слов считаем от списка cleanedText, округяем до десятых

#на выходе список с предложениями, строка с почищенным текстом, список токенов, длина полученного списка слов, список токенов после удаления стоп-слов, доля стоп-слов
    return sentences, cleanedText, tokens, 'Количество слов: '+ str(length), stopout, 'Доля стоп-слов: ' + str(sw)
text = 'Мужчинам, осужденным к лишению свободы за совершение% преступлений, предусмотренных статьями >>>>205, 2051, 2053, частью первой статьи 2054, статьей 2055, частями второй - четвертой статьи 206, статьей 208, частями второй - четвертой статьи 211, статьями 277 - 279, 281, 317, 361 настоящего Кодекса, <#отбывание части срока наказания назначается ( в тюрьме). При этом период отбывания наказания в тюрьме    после зачета времени содержания лица под @* стражей до вступления в законную силу обвинительного приговора суда должен составлять не менее одного года! Определение вида исправительного учреждения для отбывания срока наказания, оставшегося после отбытия части срока наказания в тюрьме, осуществляется! по правилам, установленным? настоящей статьей.'
#в текст в случайных местах добавлены знаки препинания
prep(text) #вызываем функцию для взятого текста
```

```python
bd = ['Номер клиента: 111, Дата: Февраль 29, 2012','Номер клиента: 112, Дата: Май 11, 2019', 'Номер клиента: 666, Дата: Май 6, 2016', 'Номер клиента: 42, Дата: Декабрь 24, 2020', 'Номер клиента: 100052, Дата: Июль 23, 2021', 'Номер клиента: 564, Дата: Февраль 1, 2005', 'Номер клиента: 98, Дата: Август 11, 2015', 'Номер клиента: 187, Дата: Апрель 14, 2013', 'Номер клиента: 87, Дата: Октябрь 29, 1999', 'Номер клиента: 88, Дата: Ноябрь 24, 2015']
```

```python
newlst = [] #новый список для столбца дат
for string in bd:
    newlst += re.findall(r'Дата: (.*)', string) #в каждом элементе списка = в каждой строке находим запись о дате, выгружая все символы после "Дата: "
print(newlst)
```

```python
with open ('text.txt', 'r', encoding = 'UTF-8') as file: #открываем файл на чтение
    lst = file.read().splitlines() #собираем данные в список строк, каждая из которых на новой строчке
    file.close #закрываем файл
resstr = '' #строка для вывода результата
for string in lst: #собираем новую строку по частям из каждой строки исходного списка
    newstr = re.findall(r'(\w+){', string)[0] + '_' + re.findall(r'=([A-Z]+)', string)[0] #во временную переменную кладем лемму (цепочку буквенных символов до {), отделитель _ и частеречный тэг (одна или несколько заглавных латинских букв после =)
    if re.search(r'[A-Z]+[=,](.+)}', string): #для слов, у которых есть "продвинутые" тэги, т. е. в майстемовском комментарии к которым есть что-нибудь после знака равно после частеречного тэга и перед }; к таким не относятся наречия, предлоги 
        newstr += '_' + re.findall(r'(\w+)\)?}', string)[0] #ищем последние буквенные цепочки перед } (мне показалось, их извлечь проще всего), так находится самый правый тэг
    resstr += newstr + ' ' #данные из временной переменной цикла сохраняем в результирующей переменной
with open ('text.txt', 'w', encoding = 'UTF-8') as fin: #с готовыми данными открываем снова тот же файл
    fin.write(resstr) #и перезаписываем его содержимое
    fin.close()
```
